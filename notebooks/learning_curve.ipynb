# PPO Learning Curve

Notebook for regenerating and inspecting the learning curve exported by the C++ PPO baseline.

```python
from pathlib import Path
import sys
from IPython.display import SVG, Markdown, display

project_root = Path.cwd()
if not (project_root / "tools" / "plot_learning_curve.py").exists() and (project_root.parent / "tools" / "plot_learning_curve.py").exists():
    project_root = project_root.parent

sys.path.insert(0, str(project_root / "tools"))

from plot_learning_curve import read_rows, render_svg

csv_path = project_root / "artifacts" / "latest" / "training_metrics.csv"
svg_path = project_root / "artifacts" / "reports" / "learning_curve.svg"

display(Markdown(f"**Project:** `{project_root}`  \\n**CSV:** `{csv_path}`  \\n**SVG:** `{svg_path}`"))

if not csv_path.exists():
    raise FileNotFoundError(
        "artifacts/latest/training_metrics.csv was not found. Run ./build/nmc train ... before opening this notebook."
    )
```

```python
rows = read_rows(str(csv_path))
render_svg(rows, str(svg_path))
display(SVG(filename=str(svg_path)))
```

```python
first = rows[0]
last = rows[-1]

print("updates:", len(rows))
print("env steps: {} -> {}".format(int(first["env_steps"]), int(last["env_steps"])))
print("avg episode return: {:.4f} -> {:.4f}".format(first["avg_episode_return"], last["avg_episode_return"]))
print("success rate: {:.4f} -> {:.4f}".format(first["success_rate"], last["success_rate"]))
print("policy loss final: {:.4f}".format(last["policy_loss"]))
print("value loss final: {:.4f}".format(last["value_loss"]))
print("entropy final: {:.4f}".format(last["entropy"]))
```
